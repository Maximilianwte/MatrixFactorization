```python
import pandas as pd
import numpy as np
import operator
```

```python
class Neighbour:
    def __init__(self):
        pass
    
    def load_data(self):
        self.data_users = pd.read_csv("../StatistaUsers10000.csv")
        self.data_tracking = pd.read_csv("../StatistaTracking10000.csv")

        return self.data_users, self.data_tracking
    
    def load_matrix(self):
        self.training_set = pd.read_pickle(f"../training_set.pkl", compression="zip")
        self.test_set = pd.read_pickle(f"../test_set.pkl", compression="zip")
        
        return self.training_set, self.test_set
    
    def calculate_neighbours(self, data, user):
        user_dict = {"user": "items"}
        user_data = data[data.idUser == user]
        
        for item in user_data.fullId:
            users_withItem = data[data.fullId == item]
            
            for user in users_withItem.idUser:
                if user not in user_dict:
                    user_dict[user] = 1
                else:
                    user_dict[user] += 1
        
        
        neighbours = list()
        
        for i in range(5):
            neighbours.append(max(user_dict.items(), key=operator.itemgetter(1))[i])
            
        return neighbours
    
    def calculate_predictions(self, user, neighbours):
        item_dict = {"item": "value"}
        user_data = data[data.idUser == user]
        user_items = data[data.idUser == user].fullId.values
        
        for neighbour in neighbours:
            neighbour_data = data[data.idUser == neighbour]
            
            for item in neighbour_data:
                if item not in item_dict:
                    item_dict[item] = 1
                else:
                    item_dict[item] += 1
                    
        
        recommendations = list()
        i = 0
        while len(recommendations) < 5:
            if i > len(user_items.items()):
                break
                
            item = max(user_dict.items(), key=operator.itemgetter(1))[i]
            if item not in user_items:
                recommendations.append(item)
            
            i += 1
            
        return recommendations
    
    def expected_percentile_ranking(self):
        accuracy = list()
        mar = 0        
        users = self.test_set.idUser.values[:80000]
        contents = self.test_set.fullId.values[:80000]
        breakpoint = 80000
        reportpoint = breakpoint if breakpoint < len(self.test_set.idUser) else len(self.test_set.idUser)
        step = 8750 if 8750 < (breakpoint/4) else 2550
        
        
        for i in range(len(users)):
            if i > breakpoint:
                break

            try:
                recommendations = self.calculate_predictions(users[i], self.calculate_neighbours(data = None, user = users[i]))
                accuracy.append(np.where(np.sort(recommendations, kind="mergesort")==contents[i])[0][0] / 50000)
            except:
                pass

            if i % step == 0:
                print(f"Solved iteration: {i}. That's about {np.round((i/reportpoint)*100,2)}%.")

        mar = np.mean(accuracy)
            
        return mar, accuracy 
```
