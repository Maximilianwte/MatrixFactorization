```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def convert(frame):
    df = frame[["idUser", "fullId"]]
    df["views"] = np.ones([len(df["fullId"]),1])
    df = df.groupby(by=["idUser", "fullId"]).sum().reset_index()
    return df
```

```python
contentType = {"Statistic": "0", "Statista-Dossier": "1", "external Report": "2", "Industry Report": "3"}
data_users = pd.read_csv("../StatistaUsers10000.csv")
training_set = pd.read_pickle(f"../training_set.pkl", compression="zip")
training_set_agg = convert(training_set)
test_set = pd.read_pickle(f"../test_set.pkl", compression="zip")
```
